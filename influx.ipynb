# Time series analysis of sensor data

Load necessary libraries:

```python
from influxdb import DataFrameClient
import pandas as pd
import plotly.express as px
import seaborn as sns
```

## influxdb

Connect to the influxDB instance with the sensor data

```python
client = DataFrameClient('192.168.1.199', 8086, '', '', 'home_assistant')
```

Query for the entity required with any time filtering.

```python
result = client.query('''
SELECT
    *
FROM "W" 
WHERE "value" < 100000
tz('Europe/London')
''')
df=result['W']
df.reset_index(inplace=True)
df.rename(columns={"index":"timestamp", "value":"W"},inplace=True)
df.shape
```

```python
fig = px.line(df, x=df.timestamp.dt.to_pydatetime(), y="W", labels = {'x':'date'},title="Power")
fig.update_xaxes(rangeslider_visible=True)
fig.show()
```
